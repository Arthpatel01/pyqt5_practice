```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [2, 4, 1, 3])
plt.show()
```

```python
#2x2 grid axis example

x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(2 * x)
y4 = np.cos(2 * x)

fig, axs = plt.subplots(2, 2, figsize=(8, 8))

axs[0, 0].plot(x, y1, color="blue")
axs[0, 0].set_title("Sin(x)")

axs[0, 1].plot(x, y2, color="orange")
axs[0, 1].set_title("Cos(x)")

axs[1, 0].plot(x, y3, color="green")
axs[1, 0].set_title("Sin(2x)")

axs[1, 1].plot(x, y4, color="red")
axs[1, 1].set_title("Cos(2x)")

plt.tight_layout()

plt.show()
```

```python
x = np.linspace(0, 2, 100)

fig, ax = plt.subplots(figsize=(5, 2.7), layout="constrained")

ax.plot(x, x, label="Linear")
ax.plot(x, x**2, label="Quadratic")
ax.plot(x, x**3, label="Qubic")

ax.set_xlabel("X Label")
ax.set_ylabel("Y Label")
ax.set_title("Simple Graph")

ax.legend()
```

```python
x = np.linspace(0, 2, 100)
plt.figure(figsize=(5, 2.7))
plt.plot(x, x, label="Linear")
plt.plot(x, x**2, label="Quadratic")
plt.plot(x, x**3, label="Qubic")

plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Simple Graph with plt")

plt.legend()

plt.show()
```

```python
def my_plotter(ax, data1, data2, param_dict):
    output = ax.plot(data1, data2, **param_dict)
    return output
```

```python
data1, data2, data3, data4 = np.random.randn(4, 100)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(5, 2.7))

my_plotter(ax1, data1, data2, {"marker": "x"})
my_plotter(ax2, data3, data4, {"marker": "o"})
```

```python
x = np.arange(len(data1))

plt.figure(figsize=(5, 2.7))
plt.plot(x, np.cumsum(data1), color="blue", linewidth=3, linestyle="--")
a, = plt.plot(x, np.cumsum(data2), color="orange", linewidth=2)
a.set_linestyle(":")
```
