```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
import os
base_dir = os.getcwd()
file_path = os.path.join(base_dir, "caliper_dup_data1.csv")

df = pd.read_csv(file_path)
df
```

```python
df.describe()
```

```python
arms_values = df.values[:, :-1]
distance_values = df.values[:, -1]

print(arms_values)
print('==========>>>')
print(distance_values)
```

```python
plt.figure(figsize=(10, 5.7))
plt.plot(distance_values, arms_values)
plt.xlabel("Distance")
plt.ylabel("Sensor Data")
plt.title("Distance VS Sensor Data")
plt.grid(True)
plt.show()
```
