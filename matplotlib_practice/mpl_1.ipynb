```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, label="Line1", color="blue", linestyle="--", marker='o')
plt.xlabel("X-axis")
plt.ylabel("Y-label")
plt.title("Test 123")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.subplot(2, 1, 1)
plt.plot(x, y, 'ro--')

plt.subplot(2, 1, 2)
plt.plot(x, y, 'g*-')

plt.subplot(2, 2, 2)
plt.plot(x, y)

plt.show()
```

```python
plt.plot(x, y)
plt.plot(y, x)
plt.show()
```

```python
data = [1, 2, 2, 3, 3, 3, 4, 4, 4.5, 5, 5, 5, 5]

plt.hist(data, bins=5, color="skyblue", edgecolor='black')
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()
```
